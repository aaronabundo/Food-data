```python
import pandas as pd
import glob
from pandasql import sqldf
import pandas as pd
import numpy as np
import csv
import seaborn as sns
import  matplotlib.pyplot as plt
import re

pysqldf = lambda q: sqldf(q, globals())
```

```python
#listing the csv files 
extension = 'csv'
csv_files = glob.glob('*.{}'.format(extension))
csv_files
```

```python
df_total = pd.DataFrame()
food_col = ['Date of fail', 
            'Product description',
            'Country of origin', 
            'Producer name',
            'Reason for fail',  
            'Result/s',
            'Units',
            'Reference']
#df_total.columns = food_col
for i in csv_files:
    df = pd.read_csv(i, header = None, encoding = 'cp1252')
    df = df.dropna(axis=0, how='all')
    #df = df.groupby(food_col)
    #df.columns = df.iloc[0]
    #df = df.drop(df.columns[~df.columns.isin(food_col)], axis=1, inplace=True)
    df_total = pd.concat([df_total, df], axis=0)

#df_total.drop(columns=df.columns[-1:], inplace=True)
#df_total.columns = df_total.iloc[0]
#df_total = df_total[1:]
#df_total.columns = df_total.columns.str.strip()
df_total
```
